1. What are Data Structures? List some of data structures and their use in real world application?

Ans: A data structure is a specific type of format used to arrange, process, retrieve, and store data. There are a number of fundamental and sophisticated forms of data structures, all of which are made to organise data for a particular function. Users may easily get the data they require and use it in the proper ways thanks to data structures.

Some of data structures and their use in real world application is as follows:

Array: Arrays are collections of identically typed elements kept in close proximity to one another in memory. They are frequently used to store data in simple and effective ways, such as lists, matrices, or tables.

Linked lists: Linked lists are made up of nodes, each of which has a reference to the following node in the list and data. They are helpful for building dynamic data structures like stacks, queues, and hash tables that demand frequent insertions and deletions.

Stacks: Stacks follow the Last-In-First-Out (LIFO) principle, where the last item added is the first one to be removed. They are used in applications that involve recursive function calls, expression evaluation, and undo/redo operations.

Queues: Queues follow the First-In-First-Out (FIFO) principle, where the first item added is the first one to be removed. They are utilized in scenarios such as job scheduling, message passing between processes, and breadth-first search algorithms.

Trees: The hierarchical data structures known as trees are made up of nodes and edges. File systems, XML/HTML parsing, decision-making techniques (such decision trees), and the organisation of hierarchical data using organisation charts are just a few areas where they find use.

Graph: A set of vertices and edges make up a graph. They are used in a variety of fields, including social networks, transportation networks, web page crawls, and optimisation issues like the shortest path algorithms.

Hash Tables: Hash tables offer effective insertion, deletion, and retrieval operations by storing data in key-value pairs. They are often utilised in associative array implementation, symbol tables, caching techniques, and databases.

Heaps: Heaps are binary trees that satisfy the heap property, where each parent node has a value greater (or smaller) than its children. They are used in priority queues, scheduling algorithms, and graph algorithms like Dijkstra's algorithm.

1. What is difference between shallow copy and deep copy?

Ans:

Shallow Copy:

i. A shallow copy creates a new object, but it references the original nested objects.

ii. Modifying a nested object in the original or the shallow copy will affect both objects.

iii. Shallow copy is performed using the copy() method or the slicing syntax [:].

iv. It is a more efficient way of copying when the objects involved are immutable or when you don't want to create copies of nested mutable objects.

Deep Copy:

i. A deep copy creates a new object and also creates new copies of all nested objects.

ii. Modifying a nested object in the original or the deep copy will not affect each other.

iii. Deep copy is performed using the deepcopy() function from the copy module.

iv. It is typically used when you want to create an independent copy of an object, including all nested mutable objects.

Example:

```python
import copy
list = [['H','O','S'],['O','E','K']]
shallow_copy = copy.copy(list)
deep_copy = copy.deepcopy(list)
list[0][0] = 'I'
print(list)
print(shallow_copy)
print(deep_copy)
```

13. Write a Python program to input basic salary of an employee and calculate its Gross salary according to following:

        Basic Salary <= 10000 : HRA = 20%, DA = 80%  

        Basic Salary <= 20000 : HRA = 25%, DA = 90%  

        Basic Salary > 20000 : HRA = 30%, DA = 95%

```python
basic_salary = float(input("Enter the BASIC SALARY of the employee: "))
if basic_salary <= 10000:
    hra_percentage = 0.2
    da_percentage = 0.8
elif basic_salary <= 20000:
    hra_percentage = 0.25
    da_percentage = 0.9
else:
    hra_percentage = 0.3
    da_percentage = 0.95
hra = basic_salary * hra_percentage
da = basic_salary * da_percentage
gross_salary = basic_salary + hra + da
print("GROSS SALARY:", gross_salary)
```

1. Write a code to implement Guess Game Guess Game
constraints / Rules

you have 5 chance to guess correct number which is choosen by computer / friend

Range which is allowed to select a number 1-50

After each guess you should give some tips / hints to user to improve their chance of winning

for example if user guess a number which is higher than actual number you should say `hint: think a low number!

if user guess a number which is lower than actual number you should say hint: think a big number!

  if user wins we will show a congrats messege and stop our game

  if any time user guess is out of limit than print a warning message saying Think in Limits 1-50 only!

  but we do not count this as a chance

```python
import random
def guess_game():
    actual_num = random.randint(1, 50)
    chances = 5
    print("Welcome to the GUESS GAME!")
    print("Guess the number (between 1 and 50).")
    print(f"You have {chances} chances to guess the correct number.")
    while chances > 0:
        user_guess = int(input("Enter your guess: "))
        if user_guess < 1 or user_guess > 50:
            print("Think in limits 1-50 only!")
            continue
        if user_guess < actual_num:
            print("Hint: Think a BIGGER number!")
        elif user_guess > actual_num:
            print("Hint: Think a SMALLER number!")
        else:
            print("CONGRATULATIONS! You guessed the correct number!")
            return
        chances -= 1
        print(f"You have {chances} chances left.")
    print("GAME OVER! You have used all your chances.")
    print(f"The actual number was {actual_num}.")
guess_game()
```

5. Write a program to sort a list without using inbuilt functions (list.sort not allowed)

i. Using Bubble Sort Algorithm

ii. Using Insertion Sort Algorithm

iii. Using Selection Sort Algorithm

```python
#i.Bubble Sort Algorithm
arr=[9,4,1,57]
print('Array before sorting: ',arr)
for i in range(0,len(arr)):
    for j in range(0,len(arr)-1-i):
        if arr[j]>arr[j+1]:
            temp=arr[j]
            arr[j]=arr[j+1]
            arr[j+1]=temp
print('Array after sorting: ',arr)
```

```python
#ii.Insertion Sort Algorithm
arr=[9,4,1,57]
print('Array before sorting: ',arr)
for i in range(1,len(arr)):
    k=arr[i]
    j=i-1
    while j>=0 and k<arr[j]:
        arr[j+1]=arr[j]
        j-=1
    arr[j+1]=k
print('Array after sorting: ',arr)
```

```python
#iii.Selection Sort Algorithm
arr=[9,4,1,57]
print('Array before sorting: ',arr)
for i in range(0,len(arr)-1):
    for j in range(i+1,len(arr)):
        if arr[i]>arr[j]:
            temp=arr[i]
            arr[i]=arr[j]
            arr[j]=temp
print('Array after sorting: ',arr)
```

```python
def rollno(roll):
    if roll>=1000 and roll<=9999:
        return True
    else:
        return False

def calcmark(marks):
    total=sum(marks)
    return total

data=[]
hightotal=0
highrollno=0
counttotal200=0

for i in range(0,10):
    print(f"For Student: {i+1}:-")
    while True:
        roll=int(input('Enter roll no: '))
        if rollno(roll):
            break
        else:
            print('Invalid Roll')

    marks=[]
    for j in range(0,3):
        marks.append(int(input(f'Marks for Subject: {j+1}: ')))
    print()

    if all(mark>=40 for mark in marks):
        totmark=calcmark(marks)

        if totmark>200:
            counttotal200+=1

        if totmark>hightotal:
            hightotal=totmark
            highrollno=roll

        data.append((roll,marks))

print('Results = ')
print(data)
print(f'No of students with total marks > 200: {counttotal200}')
print(f'Roll no of student with highest total marks: {highrollno}')
```

5. Write a program to input a number and count the digits in it. Use while loop and the program should work correctly for 0 also.

```python
num=int(input('Enter a number: '))
count=0
if num==0:
    print('No of digits: ',1)
else:
    while num:
        count+=1
        num=num//10
print('No of digits: ',count)
```

6. Write a program to enter a number and find the reverse of that number. Also display the double of the reverse number. (don’t use string magic)

```python
num=int(input("Enter a no: "))
rev=0
while num:
    a=num%10
    rev=rev*10+a
    num=num//10
print('The reversed no is: ',rev)
print('The double of reversed no: ',2*rev)
```
